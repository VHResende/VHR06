####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Código:

def dados(uf, tipo_interessado):
    url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    response = requests.get(url)
    data_json = response.json()
    if 'items' in data_json:
        dados_df = pd.DataFrame(data_json['items'])
        return dados_df
    else:
        return pd.DataFrame()
````

````python
#Testando para SP e Estado:

dados_SP_Estado = dados('SP', 'Estado')
pd.DataFrame(dados_SP_Estado)
````

````python
#Testando para RJ e Município:

dados_RJ_Municipio = dados('RJ', 'Município')
pd.DataFrame(dados_RJ_Municipio )
````

````python
# 2) Código:

dados_MG_Estado = dados('MG', 'Estado')
pd.DataFrame(dados_MG_Estado)
````

````python
# Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?
# Contando a frequência:
frequencia_MG = dados_MG_Estado['status'].value_counts()

# Mostrando a frequência:
print(frequencia_MG)
````

````python
#Retornando apenas o status 'Arquivado por decurso de prazo':

Arq_Decurso_Prazo = frequencia_MG.get('Arquivado por decurso de prazo', 0) #.get: contagem do status específico
print(f"Quantidade de solicitações 'Arquivado por decurso de prazo' para MG (Estado): {Arq_Decurso_Prazo}")
````

Podemos, também, deixar tudo num bloco de código:

````python
# Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?

# Contando a frequência:
frequencia_MG = dados_MG_Estado['status'].value_counts()

# Mostrando a frequência:
print(frequencia_MG)

#Retornando apenas o status 'Arquivado por decurso de prazo':

Arq_Decurso_Prazo = frequencia_MG.get('Arquivado por decurso de prazo', 0) #.get: contagem do status específico
print(f"Quantidade de solicitações 'Arquivado por decurso de prazo' para MG (Estado): {Arq_Decurso_Prazo}")
````

````python
# 3)  Código: Qual é o município da Bahia com mais solicitações deferidas?:

dados_BA_Municipio = dados('BA', 'Município')
pd.DataFrame(dados_BA_Municipio)
````

````python
# Contando a frequência:
frequencia_BA = dados_BA_Municipio['status'].value_counts()

# Mostrando a frequência:
print(frequencia_BA)

#Retornando apenas o status 'Deferido':

arq_deferido = frequencia_BA.get('Deferido', 0) #.get: contagem do status específico
print(f"Quantidade de solicitações deferidas para BA (Município): {arq_deferido}")
````

````python
# Visualizando os municípios:

mun_BA = dados_BA_Municipio['interessado'].value_counts()
print(mun_BA)
````

````python
#retornando o município com maior número de solicitações:

municipio_maior_def = mun_BA.idxmax()# .idxmax() obtem o índice (nome do município) de maior valor.

quantidade_def_maior_mun = mun_BA.max() # .max() obtem o maior valor do município correspondente.

print(f'Município com maior quantidade de solicitações deferidas para BA: {municipio_maior_def} - Quantidade: {quantidade_def_maior_mun}')
````

````python
#Toda a operação no mesmo código:

# Contando a frequência:
frequencia_BA = dados_BA_Municipio['status'].value_counts()

# Mostrando a frequência:
print(frequencia_BA)
print('-------------------')

#mostrando os Municípios:
mun_BA = dados_BA_Municipio['interessado'].value_counts()
print(mun_BA)
print('-------------------')

#Retornando apenas o status 'Deferido':
arq_deferido = frequencia_BA.get('Deferido', 0) #.get: contagem do status específico
print(f"Quantidade de solicitações deferidas para BA (Município): {arq_deferido}")

#retornando o município com maior número de solicitações:
municipio_maior_def = mun_BA.idxmax()# .idxmax() obtem o índice (nome do município) de maior valor.

quantidade_def_maior_mun = mun_BA.max() # .max() obtem o maior valor do município correspondente.

print(f'Município com maior quantidade de solicitações deferidas para BA: {municipio_maior_def} - Quantidade: {quantidade_def_maior_mun}')
````
